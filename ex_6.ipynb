#Project introduction

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
%matplotlib inline
```

```python
M = np.linspace(0,7,100)
gamma = 1.7
T_over_Tt = 1 / (1 + ((gamma-1)/2)*M**2)
P_over_Pt = np.pow(1 / (1 + ((gamma-1)/2)*M**2),gamma/(gamma-1))
```

```python
plt.plot(M,T_over_Tt, color="red");
plt.plot(M,P_over_Pt, color="blue");
plt.xlabel('M');
plt.ylim([0,7]);
plt.grid()
plt.legend([r"$T/T_t$", r"$P/P_t$"])
```

```python
isentropic_table = {"Mach": M, "T_over_Tt": T_over_Tt, "P_over_Pt": P_over_Pt}
isentropic_table = pd.DataFrame(isentropic_table)
isentropic_table
```

#Object Oriented Programming (OOP)

```python
class Car():
  def __init__(self, model, color, nDoor, fuel=100):
    self.model = model
    self.color = color
    self.nDoor = nDoor
    self.x = 0
    self.fuel = fuel

  def goForward(self, x=1):
    if x < 0:
      print("You can't go backward")
      return

    if self.fuel - x < 0:
      print("You don't have enough fuel")
      return

    self.x = self.x + x
    self.fuel = self.fuel - x
    print(f"You moved {x} meters")


  def move(self, x=0):

    if x < 0:
      self.x = self.x + x
      self.fuel = self.fuel - abs(x)
    elif x == 0:
      print("Car is still")
    else:
      self.goForward(x)
```

```python
class Truck(Car):
  def __super__(self):
    pass
```

```python
c1 = Car("Pride", "red", 4)
```

```python
t1 = Truck("Neisan", "blue", 2)
```

```python
c1.move(-5)
```

```python
print("Amount of Fuel: ", c1.fuel)
print("Location : ", c1.x)
```

```python
t1.move(5)
```

```python
print("Amount of Fuel: ", t1.fuel)
print("Location : ", t1.x)
```
