```python
height = input("Enter the triangle height: ")
height = int(height)
for i in range(height):
  row = ""
  for _ in range(height - (i+1)):
    row = row + " "
  for _ in range((2 * i) + 1):
    row += "#"
  print(row)
```

```python
height = input("Enter the triangle height: ")
height = int(height)
for i in range(height):
  num_space = height - (i+1)
  num_tags = (2 * i) + 1
  print(" " * num_space + "#" * num_tags)
```

```python
while True:
  height = input("Enter the triangle height: ")

  if height[0].isalpha() or height[-1].isalpha():
    print("Invalid input")
    continue
  elif "." in height:
    print("An integer number should be entered")
    continue
  elif int(height) <= 0:
    print("Number should be greater than zero")
    continue
  else:
    height = int(height)
    for i in range(height):
      num_space = height - (i+1)
      num_tags = (2 * i) + 1
      print(" " * num_space + "#" * num_tags)
    break
```
